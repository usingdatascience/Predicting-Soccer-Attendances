## Normalising a dataset

### importing libraries

```python
import csv
import pandas as pd
from csv import reader
```

### Function to find the min and max values for each column

```python
def dataset_minmax(dataset):
  minmax = list()
  for i in range(len(dataset[0])):
    col_values = [row[i] for row in dataset]
    value_min = min(col_values)
    value_max = max(col_values)
    minmax.append([value_min, value_max])
  return minmax
```

### function to rescale dataset columns to the range 0-1

```python
def normalize_dataset(df_col, min_val, max_val):
  for row in range(0,df.shape[0]):
    df.loc[row,df_col] = (df.loc[row,df_col] - min_val) / (max_val - min_val)
```
